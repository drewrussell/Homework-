```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
csv_file = "2017 Spotify .csv"
spotify_2017_df = pd.read_csv(csv_file)
spotify_2017_df.head()
```

```python
csv_file2="2018 Spotify.csv"
spotify_2018_df=pd.read_csv(csv_file2)
spotify_2018_df.head()
```

```python
spotify_2017_ms=spotify_2017_df["duration_ms"]/1000
spotify_2017_df["duration in seconds"]=spotify_2017_ms
```

```python
spotify_2018_ms=spotify_2018_df["duration_ms"]/1000
spotify_2018_df["duration in seconds"]=spotify_2018_ms
```

```python
feat_2017=spotify_2017_df[spotify_2017_df["name"].str.contains("feat")==True]
feat_2018=spotify_2018_df[spotify_2018_df["name"].str.contains("feat")==True]
```

```python
# 12 of the top 100 have a feature in 2017
# 18 of the top 100 have a feature in 2018
feat_2017.count()
feat_2018.count()
```

```python
danceabilty_2017=spotify_2017_df[spotify_2017_df['danceability']>=.7]
danceabilty_2018=spotify_2018_df[spotify_2018_df['danceability']>=.7]
```

```python
#Danceability describes how suitable a track is for dancing based on a combination of musical elements including tempo, rhythm stability, beat strength, and overall regularity. 
#A value of 0.0 is least danceable and 1.0 is most danceable.
# 55 of the 100 songs in 2017 had a danceability of >= .7
# 57 of the 100 songs in 2018 had a danceability of >= .7
danceabilty_2017.count()
danceabilty_2018.count()
```

```python
danceabilty_2017_new=danceabilty_2017[['name', 'artists', 'danceability', 'duration in seconds']]
danceabilty_2018_new=danceabilty_2018[['name', 'artists', 'danceability', 'duration in seconds']]
```

```python
danceabilty_2017_new.head()
danceabilty_2017_new.reset_index(drop=True)
danceabilty_2018_new.reset_index(drop=True)
```

```python
final_spotify = pd.merge(left=danceabilty_2017_new, right=danceabilty_2018_new, how='outer', left_on=('name','artists'),
         right_on=('name','artists'))
```

```python
final_spotify.count()
```
